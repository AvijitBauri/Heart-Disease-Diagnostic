```python
import seaborn as sns
f,ax = plt.subplots(1,2, figsize=(9,4))
df['target'].value_counts().plot.pie(explode=[0,0.1],autopct = "%1.1f%%", ax=ax[0], shadow=True)
ax[0].set_title('Target')
ax[0].set_ylabel('')
sns.countplot(x='target',data=df, ax=ax[1])
ax[1].set_title('Target')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.countplot(x='cp', data=df)
plt.xlabel('CP', fontsize=16)
plt.ylabel('Count', fontsize=16)
plt.title('Count Plot of CP Attribute', fontsize=18)
plt.xticks([0,1,2,3],["typical angina","atypical angina","non-agninal pain","asymptomatic"])
plt.xticks(rotation=45)
plt.show()
```

```python
df.columns
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x="cp", hue="target", data=df)
plt.xlabel('CP', fontsize=16)
plt.ylabel('Count', fontsize=16)
plt.title('Count Plot of CP Attribute by Target', fontsize=18)
plt.legend(title='Target')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x="fbs", hue="target", data=df)
plt.xlabel('fbs', fontsize=16)
plt.ylabel('Count', fontsize=16)
plt.title('Count Plot of fbs Attribute by Target', fontsize=18)
plt.legend(title='Target')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(9,8))
plt.xlabel('age',fontsize=16)
plt.ylabel('Count',fontsize=15)
df['trestbps'].hist(edgecolor="blue",color="green")
plt.grid(False)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style("whitegrid")

n = sns.FacetGrid(df, hue="sex", aspect=4)
n.map(sns.kdeplot, 'trestbps', shade=True)
n.add_legend(title='Sex')

plt.title('Distribution of Resting Blood Pressure (trestbps) by Sex')
plt.xlabel('Resting Blood Pressure (trestbps)')
plt.ylabel('Density')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(9,8))
plt.xlabel('age',fontsize=16)
plt.ylabel('Count',fontsize=15)
df['chol'].hist(edgecolor="black",color="skyblue")
plt.grid(False)
```
